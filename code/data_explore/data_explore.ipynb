```python
import pdfplumber
import json

def extract_ttas_tables(pdf_path):
    all_content = []
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            # 提取表格數據
            tables = page.extract_tables()
            # 提取當頁純文字（用於獲取標題與備註）
            text = page.extract_text()
            all_content.append({"text": text, "tables": tables})
    return all_content
```
